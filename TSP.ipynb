* FitnessMin: Problema de un solo objetivo que se desea minizar.
* Individual: Individuo que un array de tipo entero, también se podría haber definido como lista.
* Toolbox: Objeto caja de herramientas que nos sive para registar funciones. Se registran las siguientes funciones:
  * indices: permite generar una muestra aletoria de cidudades
  * individual: permite crear un individuo de forma aleatoria
  * population: crea un población aleatoria

```
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", array.array, typecode='i', fitness=creator.FitnessMin)

toolbox = base.Toolbox()

# Generación de una muestra del tamaño del problema
toolbox.register("indices", random.sample, range(IND_SIZE), IND_SIZE)

# Inicialización
toolbox.register("individual", tools.initIterate, creator.Individual, toolbox.indices)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
```

Función de evaluación. Siempre debe devolver una tupla porque el caso mono objetivo es un caso particular del multi objetivo

```
def evalTSP(individual):
    # distancia entre el último elemento y el primero
    distance = distance_map[individual[-1]][individual[0]]
    # distancia entre el resto de ciudades
    for gene1, gene2 in zip(individual[0:-1], individual[1:]):
        distance += distance_map[gene1][gene2]
    return distance,
```

Registramos las operaciones genéticas:
* mate: cruza dos individuos utilizando un crossover ordenado que asegura que los individuos creados sean válidos.
* mutate: muta un individuo permutando posiciones.
* select: mecanismo de selección. Se utiliza selección mediante torneo
* evaluate: función de evaluación

```
toolbox.register("mate", tools.cxOrdered) # cruce
toolbox.register("mutate", tools.mutShuffleIndexes, indpb=0.05) #mutación
toolbox.register("select", tools.selTournament, tournsize=3) # selección
toolbox.register("evaluate", evalTSP) # evaluación
```

```
random.seed(2) # ajustamos la semilla para otener siempre los mismos resultados

CXPB, MUTPB, INDIVIDUOS, NGEN = 0.8, 0.1, 300, 1000
pop = toolbox.population(INDIVIDUOS) # creamos población inicial
hof = tools.HallOfFame(1) # guardamos el mejor individuo
stats = tools.Statistics(lambda ind: ind.fitness.values) # estadística de evolución
stats.register("avg", numpy.mean)
stats.register("std", numpy.std)
stats.register("min", numpy.min)
stats.register("max", numpy.max)
algorithms.eaSimple(pop, toolbox, CXPB, MUTPB, NGEN, stats=stats, 
                        halloffame=hof)                        
print("mejor solución: ", hof)
print("fitness mejor solución: ", hof[0].fitness.values) 
```
